# 9. 데이터 선택(조건)
조건에 해당하는 데이터를 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
filt = df['이름'].isin(['채치수', '강백호'])
df[filt]
```

```python
df['키'] >= 185
```

```python
filt = df['키'] >= 185
df[filt]
```

```python
df[~filt] # filt를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학']
```

```python
df.loc[df['키']>=185, ['수학', '과학', '사회']]
```

```python
df[df['키'] >= 185][['수학','과학']]
```

### & 조건 (그리고)

```python
df[(df['키'] >= 185) & (df['학교'] == '북산고')][['수학', '과학']]
```

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

```python
filt = (df['키'] >= 185) & (df['학교'] == '북산고')
df[filt]
```

### | 조건(또는)

```python
filt = (df['키'] < 170) | (df['키'] > 200)
df.loc[filt]
```

### str 함수

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 태가 들어가는 사람을 뽑음
df[filt]
```

```python
df[~filt]
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) #SW특기가 Python이거나 Java인 사람을 뽑음
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```
